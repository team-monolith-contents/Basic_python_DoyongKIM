### **미션: 레드팀 vs 블루팀 (3)** ###

random모듈의 random.sample(리스트, 숫자)의 기능을 활용해서 24명 중에 12명을 뽑을 수 있습니다.  
그렇게 뽑은 12명은 red, 뽑히지 않은 12명은 blue에 넣으려면 어떻게 해야할까요?  
뽑은 사람은 어떻게 제거할 수 있을까요?

*  **미션** : students에 있는 값을 red, blue팀으로 랜덤하게 12명씩 나눠봅시다.

#### **힌트! 알고리즘을 먼저 생각해요**
리스트.remove(값)을 입력하면, 리스트에서 그 값을 찾아서 삭제합니다.  
중요한 것은 값을 입력해야한다는 점이예요.  
아마 더 좋은 방법도 있겠지만 먼저 알고리즘대로 만들어볼까요?  


```python
#예제 9-3: 레드팀 vs 블루팀 3

import random

students = ['찬민', '한주', '담도', '성민', '용윤', '레이', '우상', '노윤', '웅준', '윤서', '인서', '한승', '사비', '원재', '윤하', '현서', '연지', '하주', '윤지', '서현', '은예', '예지', '정현', '진현', ]
red = []
blue = []

#해당 부분 학생들에게 미제공
red = random.sample(students, 12)
for i in red:
    students.remove(i)
blue = students
#여기까지

print(red)
print(blue)
```

### **2차원 리스트** ###

#### **미션: 레드팀 vs 블루팀 (4)** ###

* **미션**: 2차원 배열을 사용하여 red팀 공격, 방어, 지휘에 각 4명씩 그리고 blue팀 공격, 방어, 지휘에 각 4명씩으로 무작위로 나눠보세요.

```python
#예제 9-5: 레드팀 vs 블루팀 (4)

import random

students = ['찬민', '한주', '담도', '성민', '용윤', '레이', '우상', '노윤', '웅준', '윤서', '인서', '한승', '사비', '원재', '윤하', '현서', '연지', '하주', '윤지', '서현', '은예', '예지', '정현', '진현', ]
red = [[], [], []]
blue = [[], [], []]

#해당 부분 학생들에게 미제공
for i in range(3):
    red[i] = random.sample(students, 4)
    for j in red[i]:
        students.remove(j)

for i in range(3):
    blue[i] = random.sample(students, 4)
    for j in blue[i]:
        students.remove(j)
#여기까지

print("red팀 공격:", red[0])
print("red팀 공격:", red[0])
print("red팀 방어:", red[1])
print("red팀 지휘:", red[2])
print("blue팀 공격:", blue[0])
print("blue팀 방어:", blue[1])
print("blue팀 지휘:", blue[2])
print("red팀 공격의 1번째 사람:", red[0][0])
print("blue팀 지휘의 3번째 사람:", blue[2][2])
```

#### **미션: 출석부 만들기** ###

* **미션** : students에 있는 학생을 알파벳 순으로 정렬하고, 같은 알파벳으로 시작하는 학생을 2차원 리스트(attendance)에 넣으세요.  
예) [[aby, ace, ate], [bee, bike, bpa]]

#### **힌트! 알고리즘을 먼저 생각해요**



```python
#예제 9-9: 출석부 만들기

students = ['chanmin', 'hanjoo', 'damdo', 'sungmin', 'yongyun', 'rei', 'woosang', 'noyoon', 'wongjun', 'yunseo', 'inseo', 'hanseoung', 'sabi', 'wonjae', 'yoonha', 'hyunseo', 'yeonji', 'hajoo', 'yoonji', 'seohyun', 'eounye', 'yeji', 'junghyun', 'jinjyun']
attendance = []

students.sort()

tmp = []

#여기서 부터
for i in students:
    if len(tmp) == 0:
        tmp.append(i)
    else:
        if i[0] == tmp[0][0]:
            tmp.append(i)
        else:
            attendance.append(tmp)
            tmp = []
            tmp.append(i)
#여기까지 가리기
            
print(attendance)
```
