## **모듈**
 
함수를 이용하여 명령어를 만들어 사용할 수 있습니다.  
이러한 함수를 모아놓은 것을 모듈이라고 부릅니다.  

프로그래머들은 서로 돕기 위해 모듈을 만들었습니다.  
그리고 모듈을 설치하면 언제든지 원할때마다 사용할 수 있습니다.  
이런 모듈은 장난감처럼 설명서만 읽고 편하게 사용할 수 있습니다.



다음 코드를 실행해봅시다.

```python
#예제 7-1: random모듈

import random

a = random.random()
b = random.randint(1,10)
c = random.choice(["안유진", "가을", "레이", "장원영", "리즈", "이서"])
d = random.sample(["셔누", "민혁", "기현", "형원", "주현", "아이엠"], 3)

print(a)
print(b)
print(c)
print(d)
```

### **time모듈**

time 모듈을 이용하면 시간을 사용할 수 있습니다.
다음 코드를 실행해봅시다.

```python
#예제 7-4: time모듈

import time as t

print("안녕하세요!")
t.sleep(1)
print("안녕하세요!")
t.sleep(10)

print("시간 재기 시작")
start = t.time()
print('현재 시간: ', start)

for i in range(10000):
    i = 1

print("시간 재기 끝")
end = t.time()
print('종료 시간: ', end)
print(end-start)
```

import time은 모듈을 사용하겠다는 코드입니다.

> sleep(숫자)는 숫자밀리초만큼 코드 실행을 중지합니다.
> time()은 1970년 1월 1일 00:00:00부터 지금까지 지나온 시간을 표시합니다.
> 이것을 잘 활용하면 코드의 실행시간을 구할 수 있습니다.

#### **미션: 구구단**

*  **미션A** : 2~9사이 무작위 정수를 2개를 a와 b에 저장하세요.
*  **미션B** : 두 숫자를 곱한 값을 물어보고, 입력받아서 정답인지 아닌지 판정하세요.
*  **미션C** : 정답이라면 정답이라는 안내와 문제를 맞추기까지 걸린 시간을 출력하세요. 예) 정답, 2초
*  **미션D** : 오답이라면 정답이 될때까지 다시 물어보고, 이때 시간은 계속 누적하여 계산하세요.
*  **미션E** : 위 구조를 반복실행하세요.

```python
# 예제 7-5: 구구단, 코드 일부 숨길 예정

import random as r
import time as t

a = r.randint(2, 9)
b = r.randint(2, 9)

print(a, '곱하기', b, '는 뭘까')
start = t.time()
answer = int(input())

while True:
    if answer == a*b:
        end = t.time()
        print("정답", end-start, '초')
        break
    else:
        print(a, '곱하기', b, '는 뭘까')
        answer = int(input())
```

#### **미션: 가위바위보**

*  **미션A** : 랜덤으로 가위, 바위, 보 중 하나를 출력하는 함수 com_choice()을 만드세요.
*  **미션B** : 가위, 바위, 보 중 하나를 입력하는 함수 hum_choice()를 만드세요.
*  **미션C** : 두 입력(a, b)를 받아서, 이긴 쪽을 출력하는 judge()함수를 만드세요.
*  **미션D** : 세 함수를 조합하여, 이긴 쪽에게 1~3점의 점수를 무작위로 주는 프로그램을 만드세요.
*  **미션E** : 이 구조를 반복하세요.

```python
#예제 7-6: 가위바위보, 코드 일부 숨길 예정

import random as r

def com_choice():
    return r.choice(["가위", "바위", "보"])

def hum_choice():
    answer = input("가위, 바위, 보 중 하나를 입력하세요")
    while True:
        if answer == "가위" or answer == "바위" or answer == "보":
            return answer
        else:
            answer = input("가위, 바위, 보 중 하나를 입력하세요")

def judge(com, hum):
    if com == hum:
        return "비김"
    if com == "가위" and hum == "바위":
        return "com"
    if com == "가위" and hum == "보":
        return "hum"
    if com == "바위" and hum == "가위":
        return "com"
    if com == "바위" and hum == "보":
        return "hum"
    if com == "보" and hum == "바위":
        return "com"
    if com == "보" and hum == "가위":
        return "hum"

com_point = 0
hum_point = 0
    
while True:
    com = com_choice()
    hum = hum_choice()
    res = judge(com, hum)
    print('컴퓨터: ', com, '사람: ', hum)
    if res == "com":
        print("컴퓨터 승")
        com_point += r.randint(1,3)    
    if res == "hum":
        print("사람 승")
        hum_point += r.randint(1,3)
    else:
        print("비김")
    print('컴퓨터: ', com_point, '사람: ', hum_point)
```
