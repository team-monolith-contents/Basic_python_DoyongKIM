```python
#  7-5: ,    

import random as r
import time as t

a = r.randint(2, 9)
b = r.randint(2, 9)

print(a, '', b, ' ')
start = t.time()
answer = int(input())

while True:
    if answer == a*b:
        end = t.time()
        print("", end-start, '')
        break
    else:
        print(a, '', b, ' ')
        answer = int(input())
```

```python
# 7-6: ,    

import random as r

def com_choice():
    return r.choice(["", "", ""])

def hum_choice():
    answer = input(", ,    ")
    while True:
        if answer == "" or answer == "" or answer == "":
            return answer
        else:
            answer = input(", ,    ")

def judge(com, hum):
    if com == hum:
        return ""
    if com == "" and hum == "":
        return "com"
    if com == "" and hum == "":
        return "hum"
    if com == "" and hum == "":
        return "com"
    if com == "" and hum == "":
        return "hum"
    if com == "" and hum == "":
        return "com"
    if com == "" and hum == "":
        return "hum"

com_point = 0
hum_point = 0
    
while True:
    com = com_choice()
    hum = hum_choice()
    res = judge(com, hum)
    print(': ', com, ': ', hum)
    if res == "com":
        print(" ")
        com_point += r.randint(1,3)    
    if res == "hum":
        print(" ")
        hum_point += r.randint(1,3)
    else:
        print("")
    print(': ', com_point, ': ', hum_point)
```
