## **조건**

"좋아. 하지만, 조건이 있어"

위와 같은 말을 사용하는 상황을 생각해봅시다.  
다음 코드를 실행해봅시다.

```python
#예제 2-1: 조건문 만나기

answer = input("대한민국의 수도는 어디입니까: ")

if answer == "서울":
    print("정답입니다")
    print("잘 아시는데요?")
```

### **참과 거짓**

answer에 받은 대답을 저장하고,  
입력받은 데이터가 서울이면 "정답입니다", "잘 아시는데요?"를 출력하는 코드입니다.

if는 영어로 만약이라는 뜻입니다.  
그리고 if 뒤에 있는 answer == "서울"은 참과 거짓을 판단하기 위한 조건식입니다.   
식이 참이면 아래 들여쓰기한 모든 코드가 실행됩니다.  

그렇다면 식의 참과 거짓은 어떻게 판별할까요?  
A == B는 A와 B가 같은지를 물어보는 조건식입니다.  
위 코드의 경우 answer에 있는 값이 서울과 같냐를 물어보고 있습니다.

따라서 answer이 서울이라면 조건식이 참이므로 들여쓰기한 코드가 실행되며  
answer이 서울이 아니면 거짓이 되어서 들여쓰기한 코드가 실행되지 않습니다.

#### **정리**

*   if 다음에 조건식을 씁니다.
*   조건문을 쓴 후에는 :(콜론)을 씁니다.
*   if가 참일 때 실행하는 코드는 4칸(스페이스바 4번 또는 tap)을 띄우고 씁니다.

다음은 if문을 이용하는 여러가지 방법입니다.

```python
#예제 2-1: 조건문 참거짓 판별

answer_1 = "사자"
answer_2 = "거위"

if answer_1 == "사자":
    print("answer이 사자였구나")
    
if answer_2 == "거위":
    print("거위도 실행될까?")

if "사자" == answer_1:
    print("위치를 바꿔도 실행될까?")

if "거위" == answer_1:
    print("위치를 바꿔지만 내용이 틀려도 실행될까?")
```

#### **코드가 어떻게 작동하는지 이해했나요?**
if문은 if뒤에 오는 조건식이 참이면 무조건 실행됩니다.
4가지 경우에서 1가지를 제외하곤 모두 실행이 됩니다.

#### **미션 : if문 기초**

if문을 처음 써본 학생이라면 실수가 많을 겁니다.  
그럴 때는 연습이 가장 좋은 해결책입니다.

*  **미션A** : 대한민국 또는 일본을 입력하면 수도를 알려주는 코드를 완성해봅시다.  
*  **미션B** : 대한민국, 일본 외에도 미국(워싱턴), 중국(베이징), 프랑스(파리), 영국(런던)도 추가해봅시다.

```python
#예제 2-3: 미션(if문 기초)

answer = input("나라를 입력하면 수도를 알려줄게: ")

if answer == "대한민국":
print("대한민국의 수도는 서울이야")

if answer = "일본"
    print("일본의 수도는 도쿄야")
```

### **if문의 활용**

if문은 주변에서 다양하게 사용됩니다.  

어떤 웹사이트에 접속할 때 아이디와 비밀번호를 판별하는 if문이 사용됩니다.  
그뿐만이 아니라, 마우스를 이용하여 화면을 클릭할 때도,

> if 클릭 상태 == "네":

와 같이 if문을 활용합니다.  
if문을 실행시키는 다양한 상황을 살펴보기 위해 다음 코드를 살펴봅시다.

```python
#예제 2-4: if문 활용

if "사자" == "사자":
    print("사자랑 사자는 같아, 실행될까?")

if "사자" == "거위":
    print("사자랑 거위는 달라, 실행될까?")

answer_1 = "짜장면"
answer_2 = "볶음밥"
answer_3 = "짜장면"

if answer_1 == answer_2:
    print("answer_1과 answer_2는 다르지, 실행될까?")

if answer_1 == answer_3:
    print("answer_1과 answer_3는 같지, 실행될까?")

if True:
    print("True는 참이라는 뜻이야")

if False:
    print("False는 거짓이라는 뜻이야")
```

#### **코드가 어떻게 작동하는지 이해했나요?**

실행되는 코드와 실행되지 않는 코드의 차이를 살펴봅시다.  
if 뒤에 오는 조건식이 참인 경우 들여쓰기한 코드는 무조건 실행됩니다.

"사자" == "사자"와 같이 항상 참인 조건문을 사용하면  
if 안에 있는 코드는 무조건 실행됩니다.

"사자" == "거위"와 같이 항상 거짓인 조건문을 사용하면  
if 안에 있는 코드는 절대 실행되지 않습니다. 

조건식에 True라고 쓰면 무조건 실행됩니다.
반대로 False라고 쓰면 무조건 실행되지 않습니다.

컴퓨터는 자신이 판단한 참과 거짓에 따라서만 움직입니다.  
우리는 참과 거짓을 판단할 있는 수 많은 코드로 컴퓨터를 제어합니다.

```python
#예제 2-5: !=의 활용

stranger = input("누구냐: ")

if stranger != "나다":
    print("너에겐 열어줄 수 없다")
```

### **부정(negative)로 물어보기**

코드를 실행하면 "나다"라고 입력하지 않아야 "너에겐 열어줄 수 없다"라고 말합니다.  
"나다"라고 입력하면, 거꾸로 아무것도 실행되지 않습니다.  
조건식에 들어가 있는 != 는 같지 않음을 물어봅니다.

> stranger != "나다" 

는 stranger에 입력된 값이 "나다"가 아니면 참이 됩니다.   
이를 다음과 같이 응용할 수 있습니다.

```python
#예제 2-6: 긍정조건문과 부정조건문

answer = input("이탈리아의 수도는: ")

if answer == "로마":
    print("정답이야")

if answer != "로마":
    print("다시 생각해봐")
```

#### **코드를 살펴봅시다**

answer에 입력된 데이터가 로마이면서 동시에 로마가 아닐 수는 없습니다.  
따라서 "정답이야" 또는 "다시 생각해봐" 중에 하나만 출력됩니다.

==와 !=가 서로 반대 관계라는 것을 이용한 것입니다.  
하지만 이 방법 말고 다른 방법도 있습니다.  
다음 코드를 살펴봅시다.

```python
#예제 2-7: else 만나기

answer = input("우리가 배우는 프로그래밍 언어의 이름은 무엇인가요: ")

if answer == "파이썬":
    print("정답이에요")
else:
    print("아직도 모른단 말이야?")
```

#### **코드가 어떻게 작동하는지 이해했나요?**

if는 영어로 만약이라고 했었지요?  
else는 그렇지 않으면 이라는 뜻입니다.

answer == "파이썬"이 참이면   
"정답이에요"가 출력되고,  

그렇지 않으면(거짓이면) "아직도 모른단 말이야?"가 출력됩니다.

앞서 !=와 ==를 동시에 사용했던 것에 비해  
else를 쓰면 코드를 굉장히 간결하게 만들고  
실행되어야 하는 코드와 그렇지 않은 코드를 간단하게 분류할 수 있습니다.

#### **미션 : if-else문 활용**

*  **미션A** : 다음 코드를 ==와 !=를 이용해서 수정해봅시다.
*  **미션B** : 다음 코드를 if-else를 이용해서 수정해봅시다.

```python
#예제 2-8: 미션(if-else 만들기)

answer = input("토끼와 거북이에서 누가 달리기를 이겼나요: ")

if answer == "거북이":
    print("제대로 기억하고 있구나")

print("아닐걸, 낮잠자고 있었을텐데")
```

### **if문 겹처 사용하기**

어떤 옷의 디자인이 다음과 같이 4종류 있습니다. 
마음에 드는 옷을 물어보려면 질문을 몇 번 해야 할까요?



다음과 같이 총 4가지의 질문을 하는 방법도 있겠지만, 더 효율적인 방법이 있습니다.

*   빨간색+민무늬를 좋아하니?
*   빨간색+줄무늬를 좋아하니?
*   파란색+민무늬를 좋아하니?
*   파란색+줄무늬를 좋아하니?

바로, 색깔을 먼저 물어보고 다음으로 민무늬인지 물어보는 것입니다(그 반대도 가능합니다.



이러면 2번의 질문으로 총 4가지 옷의 디자인을 분류할 수 있습니다.  
이러한 규칙을 이용하면 다음과 같이 코드를 작성할 수 있습니다.

다음 코드를 살펴봅시다.

```python
#예제 2-9: if문 겹쳐쓰기

answer = input("당신의 이름은 무엇입니까?: ")

if answer == "나다" :
    print("주인님 환영합니다.")
    answer = input("무슨 일로 오셨습니까: ")
    if answer == "코딩":
        print("계속 공부하시다니 존경스럽습니다.")
    else:
        print("주인님도 공부 좀 하세요.")


else:
    print("귀한 곳에 누추한 분이 오셨군요.")
    answer = input("방문 목적은 무엇입니까: ")
    if answer == "코딩":
        print("편히 공부하다 가십시오.")
    else:
        print("여긴 당신이 있을 곳이 아닙니다.")
```

#### **코드를 살펴봅시다**

꽤 복잡해보이는 코드지만 구조를 알면 어렵지 않습니다.  
if문 안에 또 if를 넣기 위해서는 4칸을 더 들여쓰기 합니다. 

그것을 이용하면 if문이 작동했을 때 또 다른 if문을 실행하게 만들 수 있습니다.
