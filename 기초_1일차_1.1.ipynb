코드가 어떻게 작동하는지 이해했나요?
이 코드는 **변수(variable)**를 하나 선언하고, 그 변수에 사용자의 입력 데이터를 저장합니다. 
변수는 변하는 것이(various) 가능하다(able)는 뜻으로 자료를 저장하는 공간입니다.

1번째 줄에 있는 **my_program**은 변수의 이름입니다.
변수의 이름은 코드를 짜는 사람이 원하는 대로 바꿀 수 있어요. 
파이썬에서 변수를 만드는 방법은 간단합니다. 변수 이름을 정하면 됩니다.

**input()**은 변수에 데이터를 입력하는 명령어 입니다. 괄호 안에는 사용자에게 보여줄 질문을 입력합니다.

**질문을 쓸 때는 반드시 따옴표를 사용해야 한다는 것을 잊지 마세요**

*   변수는 데이터를 저장하는 공간입니다.
*   변수를 만들기 위해서는 이름을 정해야 합니다.
*   input()은 변수에 데이터를 입력하는 명령어입니다.

다음 코드를 실행해봅시다.

```python
my_program = input("어떤 프로그램이 가장 마음에 드나요: ")
print("저도")
print(my_program)
print("그거 좋아해요!")
```

앞서 작성한 코드와 매우 유사한 코드입니다.

my_program이라는 변수에 마음에 드는 프로그램을 입력하고,

1줄씩 "저도", "*여러분이 입력한 값*", "그거 좋아해요!" 라고 출력이 됩니다.

print()는 무언가를 출력하는 명령어 입니다. 괄호안에 따옴표를 사용하여 출력할 글자를 입력합니다. 

변수를 출력할 때는 변수 이름만 입력하면 됩니다.

*   문자를 출력하고 싶다면, print()의 괄호 안에 따옴표를 넣어야 합니다.
*   변수를 출력하고 싶다면, print()의 괄호 안에 변수의 이름을 넣습니다.

위 주의사항을 지키지 않으면, 오류가 나거나 원하는 대로 코드가 작동하지 않을 수 있습니다.

input()명령어 안에 아무런 질문도 입력되어있지 않습니다. 그래도 코드는 정상적으로 작동합니다. 

무언가 대답을 입력하면 a에 저장이 되고, 1줄 띄고 그 다음 줄에 입력한 내용이 그대로 출력됩니다.

input() 명령어는 데이터를 입력 받을 수 있는 상태를 의미합니다. 

input() 명령어의 괄호 안에 질문을 넣어 입력 받을 데이터에 대해 설명하는 것은 프로그램을 보다 편리하게 사용하기 위한 것입니다. 

print() 명령어의 괄호 안에 아무 데이터도 입력되어 있지 않다면 아무것도 입력되지 않은 상태를 출력하는 것이기 때문에 빈 칸을 출력합니다.

```python
teacher_program = "카카오톡"
my_program = input("사진을 찍을 때 사용하는 프로그램의 이름을 입력하세요: ")

print(teacher_program, my_program)
```

techer_program과 my_program이라는 변수 2개를 사용했습니다.

techer_program에는 사용자가 입력하지 않고, 미리 카카오톡이라는 데이터를 입력해두었습니다.
