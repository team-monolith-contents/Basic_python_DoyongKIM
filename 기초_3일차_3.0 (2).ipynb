### **정수**

int() 명령어는 괄호 안 데이터를 정수로 바꿉니다.  
input() 명령어로 입력받은 데이터는 문자로 취급하기 때문에, int()명령어를 이용해서 그 값을 정수로 바꾸어주어야 숫자 계산이 가능해집니다.

다음 코드를 실행하며 몇 가지 실험을 더 해봅시다.

```python
#예제 3-4: type()으로 자료형 확인

number1 = input("숫자 하나를 입력하세요: ")
print(type(number1))
number1 = int(number1)
print(type(number1))

number2 = int(input("숫자 하나를 입력하세요: "))
print(type(number2))

print(number2+number2)
print(number2-number2)
print(number2*number2)
print((number2+number2)*number2)
print(number2+(number2*number2))
```

### **자료형**

type()명령어를 이용하면 괄호 안의 데이터가 문자, 정수인지를 확인할 수 있습니다.  
print()와 합쳐서 print(type())이라고 쓰면, type()의 괄호 안에 있는 데이터의 자료형을 확인합니다.

input()명령어로 실행한 값은 일단 문자로 취급하기 때문에 <class 'str'>라고 출력됩니다.  
자료를 저장하는 방법인 자료형(class)이 문자열(string)이라는 뜻입니다.

int()명령어로 자료형을 정수로 바꾸고 나면 <class 'int'>라고 출력됩니다.  
자료형이 정수(integer)바뀌었다는 뜻입니다.

그리고 int(input())이라는 명렁어는 int()와 input()과 합쳐서 입력받은 값을 정수로 바꾸어 저장한다는 뜻입니다.  
괄호의 순서를 조정하면 코드를 간결하게 만들 수 있습니다.

### **연산자**

+, -, *를 연산자라고 부릅니다. 연산기호와 비슷한 뜻입니다.  
더 다양한 계산을 하기 위한 연산자를 소개합니다.

*   A/B → A를 B로 나눈 값입니다(유리수로 표현)  
*   A//B → A를 B로 나눈 몫입니다(나머지 버리기)  
*   A%B → A를 B로 나눈 나머지입니다(A에서 B를 0보다 작아질 때까지 빼기)  
*   A**B → A를 B번 곱한 값입니다(A의 B제곱)  

(이미지 요청: 위 4가지 연산자는 손으로 쓴 느낌의 표 삽화로 부탁드립니다)

다음 코드를 실행해봅시다.

```python
#예제 3-6: 연산자 실습

A = 8
B = 3

print(A/B)
print(A//B)
print(A%B)
print(A**B)
```

### **반복**

컴퓨터는 복잡하고 반복적인 계산을 해주는 기계입니다.

컴퓨터에게 반복해서 계산시키는 방법을 배워봅시다. 다음 코드를 실행해보세요.

#### **for문**

for을 이용하면 반복 작업을 시킬 수 있습니다.  
"abcedfg"를 앞서 문자열(string)이라고 부른다는 것을 기억하나요?  
문자열은 문자가 줄지어 서있다는 뜻입니다.  
줄지어 서 있으면 하나씩 불러오는 것도 가능합니다.

> for i in 문자열

위와 같은 표현은 문자열에서 문자를 하나씩 떼서 변수 i에 넣으라는 뜻입니다.    
그래서 "abcdefg"에서 문자가 하나씩 출력이 되었습니다.

for문 사용 방법을 간단하게 정리하면 다음과 같습니다.

*   for 다음에 변수를 씁니다. 관습적으로 i를 많이 씁니다.  
*   변수 다음에는 in과 문자열을 쓰고, 콜론(:)을 씁니다.  
*   변수 i에 값이 저장되고, 들여쓰기한 코드가 반복 실행됩니다.  

다음은 for문을 활용하는 여러가지 방법입니다.

```python
#예제 3-9: for문 활용

for i in "월화수목금토일":
    print("오늘은" + i + "요일이야!")
    if i == "금":
        print("내일은 주말이다.")
    if i == "토":
        print("뭐하고 놀지.")
        
for i in "123456789":
    print("9번 말하기.")
```

#### **정해진 횟수만큼 반복하고 싶다면?**

for문을 활요할 때는 변수 i를 활용할 수도 있고, 활용하지 않을 수도 있습니다.

하지만 9번 같은 문장을 말하기 위해서 "123456789"와 같이 숫자열을 쓰는건 비효율적입니다.   
9번이 아니라 숫자가 더 늘어날 수도 있으니까요. 

그럴때는 range()명령어를 사용할 수 있습니다. 다음 코드를 살펴봅시다.

```python
#예제 3-10: range 기본

print("range(5)결과")
for i in range(10):
    print("안녕.")

for i in range(5):
    print(i)
```

#### **range()**

range()명령어를 사용하면 연속된 숫자를 만듭니다. 

for문과 range(10)이라는 명령어를 조합하면 10개의, 연속된 숫자를 만들어서 i에 입력합니다.  
이런 성질을 이용하면 정해진 횟수만큼 명령어를 반복할 수 있습니다.  

또한, 생성하는 숫자는 0부터 차례대로 1씩 증가하면서 입력한 갯수의 숫자를 생성합니다.  
0부터 시작하기 때문에 5는 생성하지 않는다는 것을 주의하세요.   
range(5)라는 명령어는 0부터 4까지(5보다 작은)를 생성합니다.    
그래야 5개(0, 1, 2, 3, 4)를 생성할 수 있으니까요  

다음 코드도 살펴봅시다.

```python
#예제 3-11: range 심화

print("range(2, 5)결과")
for i in range(2, 5):
    print(i)

print("range(3, 10, 2)결과")
for j in range(3, 10, 2):
    print(j)

print("range(10, 4, -1)결과")
for k in range(10, 4, -1):
    print(k)
    
a = list(range(10))
print(a)
```

#### **range()의 비밀**

range(2, 5)는 2부터 5까지 숫자를 만듭니다.   
2, 3, 4를 만듭니다. "5까지"이므로 5는 만들지 않습니다.

range(3, 10, 2)는 3부터 10까지 숫자를 만듭니다.   
2씩 더해가면서 만듭니다. 3, 5, 7, 9가 만들어집니다.

range(10, 4, -1)은 10부터 4까지 숫자를 만듭니다.  
-1씩 더해가면서 즉 1씩 빼가면서 만듭니다. 10, 9, 8, 7, 6, 5가 만들어집니다. "4까지"이므로 4는 만들지 않습니다.
