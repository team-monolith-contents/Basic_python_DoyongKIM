### **오류 해결하기**

파이썬은 엔트리와는 다르게 입력을 실수하면 아예 실행이 안 되기도 합니다.  
코드를 수정하지 않고 그대로 실행하면 나오는 메시지는 다음과 같습니다.

>

#### **오류 메시지를 살펴봅시다.**

**----> 2 print(선생님은)**  
는 2번째 줄에서 오류가 발생했다는 뜻입니다.

**NameError: name '선생님은' is not defined**  
오류의 종류는 NameError, 이름과 관련 된 오류입니다.   
'선생님은'이라는 이름의 변수가 정의되지(defined) 않았다(not)다는 것이지요.

명령어 print()에서 따옴표를 쓰지 않은 글은 변수의 이름으로 취급합니다.  
변수를 사용하기 위해서는 반드시 변수의 이름을 먼저 선언해주어야 합니다.

심부름을 할 때 모르는 물건을 사올 수 없는 것처럼,  
출력을 할 때도 모르는 변수는 출력할 수 없습니다.

다음 코드를 실행해봅시다.

```python
#예제 1-5: 명령어의 빈칸

a = input()
print()
print(a)
```

#### **코드가 어떻게 작동하는지 이해했나요?**

input()명령어의 괄호 안이 비어있지만 코드는 정상적으로 작동합니다.  

아무거나 입력하고 엔터키를 누르면  
그 값이 **a** 에 저장이 되고  
1줄을 띄운후  
그 다음 줄에 **a**의 값이 출력됩니다.  

input() 명령어를 사용할 때 입력 받을 데이터에 대해 설명하는 것은  
인간을 위해서 프로그램을 보다 편리하게 사용하기 위한 것입니다.  
괄호 안에 꼭 질문이 들어갈 필요는 없습니다.

print() 명령어의 괄호 안이 비어 있으면 아무 것도 아닌 값을 출력합니다.  
따라서 빈 칸을 출력하게 되어 1줄이 띄어집니다.  

다음 코드를 살펴봅시다.

```python
#예제 1-6: 두 변수 출력

teacher_program = "카카오톡"
my_program = input("사진을 찍을 때 사용하는 프로그램의 이름을 입력하세요: ")

print(teacher_program, my_program)
```

### **변수에 데이터 덮어쓰기**

**techer_program**과 **my_program**이라는 변수 2개를 사용합니다.  
techer_program에는 사용자가 입력하지 않고, 미리 카카오톡이라는 데이터를 입력해두었습니다.    

print(변수, 변수)는 공백을 두고 두 변수를 출력하라는 뜻입니다.   
실행하면 teacher_program과 my_program이 순서대로 출력되는 것을 관찰할 수 있습니다.  
그런데 이미 변수에 데이터가 들어 있는 상태에서 다른 데이터를 입력하면 어떻게 될까요?  


다음 코드를 살펴봅시다.

```python
#예제 1-7: 데이터 덮어쓰기

my_program = input("엑셀을 제외하고 다른 단어를 입력하세요: ")
my_program = "엑셀"

print("내가 사용해본 프로그램에는", my_program, "가 있어.")
```

#### **코드가 어떻게 작동하는지 이해했나요?**

1번째 줄에서 변수 my_program에 입력을 받지만,  
다음 줄에서 my_program에 엑셀을 입력하라는 코드가 있습니다.  
따라서, 사용자가 무엇을 입력하던 5번째 줄에서 my_program을 출력하면 "엑셀"이 나옵니다.

파이썬의 코드는 위에서부터 아래로 실행됩니다.  
이런 규칙을 간과하면 의도한 것과 다르게 코드가 작동할 수 있어요.
