## **파일**

pdf, docx, hwp, txt로 끝나는 종류의 파일을 문서 파일이라고 부릅니다.  
메모장이나 워드프로세스를 이용하여 이러한 문서 파일을 열어본적이 있을겁니다.  

이러한 문서 파일은 데이터로 이루어져있습니다.  
그리고 파이썬으로도 이러한 파일을 직접 다룰 수 있습니다.  

다음 코드를 살펴봅시다.

```python
#예제 7-1: file기록

file = open("name.txt", "w")
file.write("정국")
file.write("알엠")
file.write("뷔")
file.write("\n")
file.write("방탄소년단")
file.close()
```

### **텍스트 파일** ###

txt는 메모장으로 열 수 있는 파일입니다.  
위 코드를 실행하면 화면 왼쪽의 폴더 name.txt라는 파일이 생깁니다.
더블클릭해서 파일을 실행해보세요.  

#### **코드가 어떻게 작동하는지 이해했나요?** ####

open("name".txt", "w")는 name이라는 txt파일을 생성하고, 변수 file에 Write(쓰는) 명령어입니다.  
name의 이름을 바꾸면 다른 이름의 텍스트 파일이 생성됩니다.

file.write()는 file에 괄호 안의 데이터를 입력하라는 뜻입니다.  
"\n"는 줄을 띄워서 입력하는 문자열입니다.

file.close()는 file로 열었던 파일을 닫는다는 뜻입니다.

다음 코드를 실행해봅시다.

```python
#예제 7-2: file오픈

file = open("name.txt", "r")
text = file.read()   
print(text)
file.close()

file = open("name.txt", "r")
text = file.readlines()
print(text)
file.close()
```

#### **코드가 어떻게 작동하는지 이해했나요?** ####

open("name".txt", "r")는 name이라는 txt파일을 read(읽기)모드로 여는 명령어입니다.  
읽기모드에서는 파일의 내용을 읽어올 수 있습니다.

read()는 데이터를 그대로 모두 읽어옵니다.
readlines()는 데이터를 줄별로 읽어와서 배열로 읽어옵니다.

#### **미션: 노래방 기계**

*  **미션A** : name.txt파일을 직접 열어서 인터넷에서 찾은 노래가사를 입력해보세요. (줄구분 필수)
*  **미션B** : time()모듈의 sleep()를 이용하여, 노래 가사를 천천히 1줄씩 출력하세요.
*  **미션C** : 1줄씩 출력할 때, \n에 때문에 1줄이 띄워지는 현상을 수정해보세요.



```python
#예제 7-3: 미션(노래방 기계)

import time as t

file = open("name.txt", "r")
text = file.readlines()

print(text)

for i in text:
    print(i)
    t.sleep(1)
   
file.close()
```

#### **덮어쓰기** ####

파일에 새로운 내용을 이어서 쓰려면 어떻게 해야할까요?  
다음 코드를 **여러 번** 실행해봅시다.

```python
#예제 7-4: 덮어쓰기

f = open("name.txt", "a")
f.write("참외")
f.write("멜론")
f.write("포도")

f = open("name.txt", "r")
text = f.read()    
print(text)
f.close()
```

#### **코드가 어떻게 작동하는지 이해했나요?** ####

file대신에 f를 사용할 수도 있습니다.  
file의 내용이 저장되는 변수가 f로 바뀝니다.

open("name".txt", "a")는 name이라는 txt파일을 append(추가)모드로 여는 명령어입니다.  
수정 모드에서는 뒤에 파일이 쭉 이어서 써집니다. 
f.writelines([])는 리스트 안의 요소를 

따라서, 반복 실행을 하면 내용이 계속 늘어나는 것입니다.

#### **코드가 어떻게 작동하는지 이해했나요?** ####

모드 | 용어 | 의미 | 내용 | 주요 코드
--- | --- | --- | --- | ---
r | read | 읽기 | 파일의 내용을 읽어 온다 | f=open("a.txt","r"), f.read(), f.readlines()
w | write | 쓰기 | 파일을 생성하고 입력 내용을 모두 지운뒤 처음부터 쓴다. | f=open("a.txt","w"), f.write()
a | append | 추가하기 | 파일을 생성하고 입력 내용을 기존 파일 내용의 뒤에 추가한다. | f=open("a.txt","a"), f.write()

```python
# 예제 7-6: calendar모듈

import calendar as c

print(c.calendar(2023))
print(c.monthcalendar(2023, 7))

print(c.weekday(2023, 7, 1))
print(c.weekday(2023, 7, 2))
print(c.weekday(2023, 7, 3))
print(c.weekday(2023, 7, 4))
print(c.weekday(2023, 7, 5))
print(c.weekday(2023, 7, 6))
print(c.weekday(2023, 7, 8))
```

#### **코드가 어떻게 실행되는지 이해했나요?**

c.monthcalendar(년, 월)는 연도의 월 달력을 불러옵니다.  
c.weekday(년, 월, 일)는 해당 날짜의 요일을 불러옵니다.
> 0==월, 1==화, 2==수, 3==목, 4==금, 5==토, 6==일
